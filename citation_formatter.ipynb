Build a basic PubMed API Scraper

Script will fetch medical article abstracts from PubMed using public API. 

[Analyze Scientifici publications with E-utilities and python](https://towardsdatascience.com/analyze-scientific-publications-with-e-utilities-and-python-56f76de22959/)


```python
# Use NCBI E-utilities (API) using BioPython library
# Might need to install biopyathon package > pip install biopython
from Bio import Entrez

# Always tell NCBI who you are 
Entrez.email = "ryancytso@gmail.com"

# Define search query
search_term = "cancer therapy"

# Use eSearch to find PMIDs (set max number of papers to 5)
handle = Entrez.esearch(db="pubmed", term=search_term, retmax=1)
record = Entrez.read(handle)
handle.close()

# Save article IDs
id_list = record["IdList"]
print(id_list)
```

```python
# use eFetch to get all the details of those articles from id_list
handle = Entrez.efetch(db="pubmed", id=id_list, rettype="medline", retmode="xml")
print(handle.read())
handle.close()
```
